```python
# Imporing required library
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
from posixpath import join
list_content = []
listed_link = []
headline = []
time = []
hoax = []

def web_scraping(keywords, numbers_of_pages):
  url = f'https://www.detik.com/search/searchall?query={keywords}&siteid=2&sortby=time&page={numbers_of_pages}'
  response = req.get(url).text
  soup = bs(response, 'html.parser')
  pages = soup.find('div', class_ = 'list media_rows list-berita')
  articels = pages.find_all('article')

  for i in articels:
    # get the link of the content
    link = i.find('a')['href']

    # checking for duplicated data
    if link not in listed_link:
      urls_listed_link = link
      req_article_link = req.get(urls_listed_link).text
      soup_article_link = bs(req_article_link,'html.parser')

      num_pages = soup_article_link.find('div', class_='detail__anchor mgr-4')
      if num_pages is not None:
        # get the content
        content = soup_article_link.find('div',class_='detail__body-text itp_bodycontent')
        if content is not None:
          # print(link)
          listed_link.extend([link])

          # get the headline
          header = soup_article_link.find('h1', class_='detail__title').text
          # pre cleaning header text
          header = header.split()
          header = ' '.join(header)
          headline.extend([header])
          # print(header)

          # get the time
          timeline = soup_article_link.find('div', class_='detail__date')
          time.extend(timeline)

          x = content.find_all('p')
          y = [y.text for y in x ]

          join_xy = ''.join(y)
          # text precleaning
          join_xy = join_xy.split("Selengkapnya halaman selanjutnya. ")[0]
          join_xy = join_xy.replace("ADVERTISEMENT", "")
          join_xy = join_xy.replace("SCROLL TO CONTINUE WITH CONTENT", "")
          join_xy = join_xy.split()
          join_xy = ' '.join(join_xy)
          # print(join_xy)

          #get the max num on the pages
          num_pages = num_pages.find_all('a')
          num_pages = int(num_pages[-1].get_text())

          for num in range(2, num_pages+1):
            contents_array = []
            # get content on different number pages
            pages_link = urls_listed_link+'/'+str(num)
            req_article_link_pages = req.get(pages_link).text
            soup_article_link_pages = bs(req_article_link_pages,'html.parser')
            content_article_update = soup_article_link_pages.find('div',class_='detail__body-text itp_bodycontent')
            x_article_update = content_article_update.find_all('p')
            y_article_update = [y_upd.text for y_upd in x_article_update ]
            join_xy_article_update = ''.join(y_article_update)
            join_xy_article_update = join_xy_article_update.split("Selengkapnya halaman selanjutnya. ")[0]
            join_xy_article_update = join_xy_article_update.split("Baca halaman selanjutnya.")[0]
            join_xy_article_update = join_xy_article_update.replace("ADVERTISEMENT", "")
            join_xy_article_update = join_xy_article_update.replace("SCROLL TO CONTINUE WITH CONTENT", "")
            join_xy_article_update = join_xy_article_update.split()
            join_xy_article_update = ' '.join(join_xy_article_update)
            contents_array.extend([join_xy_article_update])
            # print(join_xy_article_update)
          # print(join_xy+' '.join(contents_array)+'\n')
          list_content.extend([join_xy+' '.join(contents_array)])
          # add status
          hoax.append(0)
          print('Success adding data')
          print(f'length for content :{len(list_content)}\nlength for headline :{len(headline)}\nlength for time :{len(time)}\nlength for hoax :{len(hoax)}\nlength for link:{len(listed_link)}\n')

        else:
          pass

      else:
        # checking the error contet
        content = soup_article_link.find('div',class_='detail__body-text itp_bodycontent')
        if content is not None:
          # print(urls_listed_link)
          listed_link.extend([link])
          # get the header
          header = soup_article_link.find('h1', class_='detail__title').text
          # pre cleaning header text
          header = header.split()
          header = ' '.join(header)
          headline.extend([header])
          # print(header)

          # get the time
          timeline = soup_article_link.find('div', class_='detail__date')
          time.extend(timeline)

          x = content.find_all('p')
          y = [y.text for y in x ]
          join_xy = ''.join(y)
          # text precleaning
          join_xy = join_xy.split("Selengkapnya halaman selanjutnya. ")[0]
          join_xy = join_xy.replace("ADVERTISEMENT", "")
          join_xy = join_xy.replace("SCROLL TO CONTINUE WITH CONTENT", "")
          join_xy = join_xy.split()
          join_xy = ' '.join(join_xy)
          # print(join_xy+'\n')
          list_content.extend([join_xy])
          hoax.append(0)
          print('Success adding data')
          print(f'length for content :{len(list_content)}\nlength for headline :{len(headline)}\nlength for time :{len(time)}\nlength for hoax :{len(hoax)}\nlength for link:{len(listed_link)}\n')
        else:
          pass
    else:
      print('Skipping for duplicated data\n')
```

```python
keywords = ['palestina','israel','hamas','zionis']
numbers = 10
for keyword in keywords:
  for number in range(1, numbers+1):
    web_scraping(keyword, number)
```
