```python
df.dropna(subset=['content'], inplace=True)
```

```python
df.isna().sum()
```

```python
import re
```

```python
def extract_date(date_string):
    pattern = r'(\d{1,2})\s\w{3}\s\d{4}'
    match = re.search(pattern, date_string)
    if match:
        return match.group(1)
    return None

# Mengekstrak data tanggal
df['date'] = df['time'].apply(lambda x: extract_date(x))
```

```python
df['date'] = df['date'].astype(int)
```

```python
df.isna().sum()
```

```python
df.rename(columns={'bulan (full)':'month',
                   'tahun':'year'},inplace=True)
```

```python
df = df[['headline','date','month','year','content','link','hoax']]
```
