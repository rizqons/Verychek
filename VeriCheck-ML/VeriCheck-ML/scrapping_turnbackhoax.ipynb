```python
# Imporing required library
import requests as req
from bs4 import BeautifulSoup as bs
```

Dataset hoax yang digunkan [disini](https://turnbackhoax.id/tentang-kami/).

```python
list_content = []
listed_link = []
headline = []
time = []
hoax = []

def web_scraping(keyword):

  # get the max number page
  urls = f'https://turnbackhoax.id/page/1/?s={keyword}'
  response = req.get(urls).text
  soup = bs(response, 'html.parser')
  pages = soup.find('div', class_ = 'nav-links')
  page = pages.find_all('a')
  num_page  = int(page[-2].get_text())

  # renumber the pages
  for num in range (1,num_page+1):
    urls = f'https://turnbackhoax.id/page/{num}/?s={keyword}'
    response = req.get(urls).text
    soup = bs(response, 'html.parser')

    find_content = soup.find('div',id='main-content')
    list_article = find_content.find_all('article')

    for i in list_article:
      link = i.find('a')['href']
      if link not in listed_link:
        # get the link to the list
        listed_link.extend([link])

        req_article_link = req.get(link).text
        soup_article_link = bs(req_article_link,'html.parser')

        # getting the content
        content = soup_article_link.find('div',class_='entry-content mh-clearfix')
        x = content.find_all('p')
        y = [y.text for y in x ]
        join_xy = ''.join(y)

        # getting the headline
        header = soup_article_link.find('h1', class_ = 'entry-title').text

        # getting the time
        timeline = soup_article_link.find('span', class_ = 'entry-meta-date updated')
        timeline = timeline.find('a').text

        # adding data to the list
        list_content.extend([join_xy])
        headline.extend([header])
        time.extend([timeline])
        hoax.append(1)
        print('Succes adding data')
        print(f'length for content :{len(list_content)}\nlength for headline :{len(headline)}\nlength for time :{len(time)}\nlength for hoax :{len(hoax)}\nlength for link data:{len(listed_link)}\n')
      else:
        print('Skipping duplicated data\n')
```

```python
keywords = ['israel', 'palestina', 'zionis', 'hamas']
for keyword in keywords:
  web_scraping(keyword)
```

```python
import csv
```

```python
data = zip(headline, time, list_content, listed_link, hoax)

# Menyimpan data ke dalam file CSV
with open('data_turnbachoax_dirty.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['headline', 'time', 'content','link','hoax'])  # Menulis header
    writer.writerows(data)  # Menulis data
```
